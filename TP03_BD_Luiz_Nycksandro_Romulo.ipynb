#### **Tarefa 13 - Comparações de operadores de agregação.**

Considere as seguintes consultas em SQL, sobre o atributo VOTES, as quais são equivalentes:

* SELECT title FROM movie WHERE votes >= (SELECT MAX(votes) FROM movie); 
* SELECT title FROM movie WHERE votes >= ALL (SELECT votes FROM movie) ;

>**a)** Apresente o resultado do comando EXPLAIN ANALYZE sobre as duas consultas acima  
>**b)** Existe alguma diferença entre os planos de consultas? Qual das duas é mais eficiente? Explique?

**O que entregar:** Relatório com as respostas das questões.


```python
#a)

#1) Consulta:
EXPLAIN ANALYZE
SELECT title FROM movie WHERE votes >= (SELECT MAX(votes) FROM movie);

#2) Consulta:
EXPLAIN ANALYZE
SELECT title FROM movie WHERE votes >= ALL (SELECT votes FROM movie);
```

#### **Resposta da letra a)**

##### Para a primeira consulta: Retorna 1 linha que satisfaz a condição

Index Scan using movie_votes on movie  (cost=0.61..35.37 rows=615 width=16) (actual time=0.174..0.190 rows=1 loops=1)  
>Index Cond: (votes >= $1)  
>InitPlan 2 (returns $1)  
    >> -> Result  (cost=0.32..0.33 rows=1 width=4) (actual time=0.146..0.157 rows=1 loops=1)  
        >>>InitPlan 1 (returns $0)
            >>>> -> Limit  (cost=0.28..0.32 rows=1 width=4) (actual time=0.135..0.143 rows=1 loops=1)
                >>>>> ->  Index Only Scan Backward using movie_votes on movie movie_1  (cost=0.28..76.55 rows=1844 width=4) (actual time=0.131..0.133 rows=1 loops=1)   
                    >>>>>> Index Cond: (votes IS NOT NULL)  
                    >>>>>> Heap Fetches: 0  

 Planning Time: 73.728 ms  
 Execution Time: 0.352 ms

##### Para a segunda consulta: Retorna 1 linha que satisfaz a condição

Seq Scan on movie  (cost=0.00..43620.99 rows=922 width=16) (actual time=15.678..23.853 rows=1 loops=1)  
   >Filter: (SubPlan 1)  
   >Rows Removed by Filter: 1843  
   >SubPlan 1  
    >> ->  Materialize  (cost=0.00..42.66 rows=1844 width=4) (actual time=0.002..0.008 rows=2 loops=1844)  
        >>>   ->  Seq Scan on movie movie_1  (cost=0.00..33.44 rows=1844 width=4) (actual time=0.006..4.026 rows=1844 loops=1)  

 Planning Time: 0.142 ms  
 Execution Time: 24.045 ms  

#### **Resposta da letra b)**

Sim, existe. Ambas utilizam estratégias bem diferentes:
* Primeira Consulta: Utiliza duas subconsultas em índices. A primeira executa Index Only Scan Backward, que acessa o índice e percorre no sentido inverso até encontrar o valor máximo de votos, guardando na váriavel $0 e disponibilizando para o InitPlan2 como a váriavel $1. Após isso, é executado um Index Scan para encontrar os votos maiores que o encontrado no primeiro Index Scan.
* Segunda Consulta: Primeiramente, é executado um Seq Scan como filtro, onde é varrido a tabela inteira e o resultado é salvo temporariamente como movie_1 atráves do Materialize, apoś isso é executado outro Seq Scan para varrer a tabela movie inteira e para cada linha lida, é utilizada o movie_1 para a aplicação do filtro.

A primeira consulta é muito mais eficiente do que a segunda (cerca de 68 vezes mais rápido), tal afirmação pode ser comprovada observando as estatísticas cedidas na consulta. Tanto o custo estimado pelo otimizador, quanto o tempo de execução são bem menores na primeira consulta, além disso, a estratégia adotada na consulta 1 irá acessar menos a mémoria secundária que na consulta 2. Porém, é válido ressaltar, o custo associado a verificação de cada plano, sendo o custo de verificação da segunda consulta bem menor que o da primeira (cerca de 519 vezes menor).
